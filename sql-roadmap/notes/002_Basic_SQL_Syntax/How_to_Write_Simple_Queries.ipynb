## <a id='toc1_4_'></a>[Что такое SQL?](#toc0_)
SQL — это язык программирования, который позволяет управлять реляционными базами данных и манипулировать ими. Обычно его произносят как «сиквел». SQL — важный инструмент для компаний, которым необходимо регулярно получать доступ к большим массивам данных и анализировать их. SQL позволяет извлекать определённые данные с помощью запроса, обновлять существующие данные, вставлять новые данные, удалять данные и делать многое другое.


С помощью SQL вам не нужно скачивать и открывать огромную таблицу Excel, чтобы получить нужные ответы.

Вы можете задавать такие вопросы, как «Кто из покупателей за последние шесть месяцев приобретал красный комбинезон?», а SQL извлечёт данные из вашей базы и вернёт их вам без необходимости вручную просматривать CSV-файл.

## <a id='toc1_5_'></a>[Зачем использовать SQL?](#toc0_)
SQL — полезный инструмент для компаний, которые работают с данными (подсказка: большинство из них так и делают). Вот несколько примеров и причин, по которым вам стоит освоить SQL:

* Ваши данные в SQL-базе защищены лучше, поскольку пользователям сложнее случайно удалить или повредить их, чем в таблице Excel
* SQL позволяет управлять наборами данных, содержащими тысячи записей
* SQL позволяет нескольким пользователям беспрепятственно получать доступ к одной базе данных
* Ролевая авторизация позволяет контролировать доступ к конфиденциальным данным
* SQL обеспечивает мощную визуализацию данных
* SQL обеспечивает целостность данных, поэтому ваши данные всегда точны и непротиворечивы

## <a id='toc1_6_'></a>[Иерархия базы данных SQL](#toc0_)
База данных SQL — это реляционная база данных, то есть данные в ней структурированы в виде таблиц, связанных друг с другом на основе заранее определённых отношений.

Информация в базе данных SQL структурирована иерархически, подобно генеалогическому древу. Это означает, что элементы верхнего уровня имеют более широкую область применения и разветвляются на несколько более конкретных подуровней.

В контексте SQL верхний уровень — это сервер базы данных, также называемый экземпляром. В вашем экземпляре хранятся все ваши данные. В экземпляре может быть несколько баз данных, каждая из которых содержит данные, организованные по какой-либо общей категории.

База данных разбита на таблицы. В таблице хранятся фактические данные. На уровне таблицы данные организованы по столбцам и строкам и размещены в полях, почти как в электронной таблице Excel.

Как писать SQL-запросы

Чтобы научиться писать SQL-запросы, давайте рассмотрим следующий вопрос:

Кто эти рыжеволосые люди из Массачусетса, родившиеся в 2003 году?

## <a id='toc1_7_'></a>[Использование команды SELECT](#toc0_)
SELECT выбирает поля, которые должны отображаться на диаграмме. Это конкретная информация, которую вы хотите получить из базы данных. В приведенном выше примере мы хотим найти людей, которые соответствуют остальным критериям.

## <a id='toc1_8_'></a>[Используя команду FROM](#toc0_)
FROM, вы точно определяете таблицу, из которой вы хотите извлечь данные.

В предыдущем разделе мы узнали, что для каждого из шести штатов Новой Англии существует шесть таблиц: people_connecticut, people_maine, people_massachusetts, people_newhampshire, people_rhodeisland и people_vermont.

Поскольку мы ищем людей конкретно в Массачусетсе, мы будем извлекать данные из этой конкретной таблицы.

Вот наш SQL-запрос:

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts;
```

## <a id='toc1_9_'></a>[Команда WHERE](#toc0_)
WHERE позволяет отфильтровать запрос, чтобы он был более конкретным. В нашем примере мы хотим отфильтровать наш запрос так, чтобы он включал только людей с рыжими волосами, родившихся в 2003 году. Давайте начнем с фильтра рыжих волос.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red';
```

параметр hair_color мог бы быть частью вашей первоначальной инструкции SELECT, если бы вы хотели просмотреть всех жителей Массачусетса и их цвет волос. Но если вы хотите отфильтровать, чтобы увидеть только людей с рыжими волосами, вы можете сделать это с помощью инструкции WHERE.

## <a id='toc1_10_'></a>[Используя команду BETWEEN](#toc0_)
Помимо equals (=), BETWEEN - это еще один оператор, который можно использовать для условных запросов. Оператор BETWEEN имеет значение true для значений, которые находятся между указанными минимальными и максимальными значениями.

В нашем случае мы можем использовать BETWEEN для извлечения записей за определенный год, например, за 2003.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE birth_date BETWEEN '2003-01-01' AND '2003-12-31';
```

## <a id='toc1_11_'></a>[С помощью команды AND](#toc0_)
AND вы можете добавить дополнительные критерии в свой запрос WHERE. Помните, что мы хотим выполнить фильтрацию по людям с рыжими волосами, а не по людям, родившимся в 2003 году. Поскольку наш оператор WHERE учитывает критерии рыжих волос, как мы можем выполнить фильтрацию по конкретному году рождения?

Вот где используется оператор AND. В данном случае оператор AND является свойством даты, но это не обязательно должно быть так. (Примечание: Уточните формат ваших дат у вашей команды разработчиков, чтобы убедиться в их правильности.)

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31';
```

## <a id='toc1_12_'></a>[С помощью команды OR](#toc0_)
OR также можно использовать с оператором WHERE. При использовании AND оба условия должны быть выполнены, чтобы они отображались в результатах (например, цвет волос должен быть рыжим и человек должен быть 2003 года рождения). При использовании OR в результатах поиска должно выполняться любое из условий (например, цвет волос должен быть рыжим или человек должен быть 2003 года рождения).

Вот как выглядит оператор OR в действии.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
   OR birth_date BETWEEN '2003-01-01' AND '2003-12-31';
```

## <a id='toc1_13_'></a>[С помощью команды NOT](#toc0_)
NOT используется в операторе WHERE для отображения значений, в которых указанное условие не соответствует действительности. Если бы мы хотели найти всех жителей Массачусетса, у которых нет рыжих волос, мы могли бы использовать следующий запрос.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE NOT hair_color = 'red' ;
```

## <a id='toc1_14_'></a>[С помощью команды ORDER BY](#toc0_)
Вычисления и организация также могут быть выполнены в рамках запроса. Именно здесь используются функции ORDER BY и GROUP BY. Сначала мы рассмотрим наши SQL-запросы с функциями ORDER BY, а затем GROUP BY. Затем мы кратко рассмотрим разницу между ними.

Предложение ORDER BY позволяет выполнять сортировку по любому из полей, которые вы указали в инструкции SELECT. В данном случае давайте упорядочим по фамилии.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31'
ORDER BY last_name;
```

## <a id='toc1_15_'></a>[Использование команды GROUP BY](#toc0_)
Функция GROUP BY аналогична функции ORDER BY, но объединяет похожие данные. Например, если в ваших данных есть дубликаты, вы можете использовать функцию GROUP BY для подсчета количества дубликатов в ваших полях.

```python
%%sql

SELECT last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31'
GROUP BY last_name;
```

## <a id='toc1_17_'></a>[Использование функции ОГРАНИЧЕНИЯ](#toc0_)
Выполнение ваших запросов может занять много времени, в зависимости от объема данных, хранящихся в вашей базе данных. Это может привести к разочарованию, особенно если вы допустили ошибку в своем запросе и теперь вам нужно подождать, прежде чем продолжить. Если вы хотите протестировать запрос, функция LIMIT позволяет ограничить количество получаемых результатов.

Например, если мы подозреваем, что в Массачусетсе у тысяч людей рыжие волосы, мы можем захотеть протестировать наш запрос с использованием LIMIT, прежде чем запускать его полностью, чтобы убедиться, что мы получаем нужную информацию. Допустим, например, мы хотим видеть в нашем результате только первые 100 человек.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31'
ORDER BY last_name
LIMIT 100;
```

## <a id='toc1_18_'></a>[С помощью команды INSERT INTO](#toc0_)
Помимо извлечения информации из реляционной базы данных, SQL также можно использовать для изменения содержимого базы данных.

Разумеется, для изменения данных вашей компании вам потребуется разрешение. Но на случай, если вам когда-нибудь придется управлять содержимым базы данных, мы поделимся некоторыми запросами, которые вам следует знать.

Во-первых, это инструкция INSERT INTO для ввода новых значений в вашу базу данных.

Если мы хотим добавить нового пользователя в таблицу Massachusetts, мы можем сделать это, сначала указав название таблицы, которую мы хотим изменить, и поля в таблице, которые мы хотим добавить.

Затем мы записываем значение для каждого соответствующего значения, которое мы хотим добавить.

```python
%%sql

INSERT INTO people_massachusetts (address_city, address_state, address_zip, hair_color, age, first_name, last_name)
VALUES ('Cambridge', 'Massachusetts', '02139', 'blonde', 32, 'Jane', 'Doe');
```

В качестве альтернативы, если вы добавляете значение в каждое поле таблицы, вам не нужно указывать поля. Значения будут добавлены в столбцы в том порядке, в котором они указаны в запросе.

```python
%%sql

INSERT INTO people_massachusetts VALUES
(999, 'Jane', 'Doe', 32, 'blonde', '1989-09-21', 'Cambridge', 'Massachusetts', '02139');
```

Если вы хотите добавить значения только в определенные поля, вы должны указать эти поля. Допустим, мы хотим вставить запись только с именем, фамилией и адресом_state — мы можем использовать следующий запрос.

```python
%%sql

INSERT INTO people_massachusetts (first_name, last_name, address_state)
VALUES ('Jane', 'Doe', 'Massachusetts');
```

## <a id='toc1_19_'></a>[Используя команду UPDATE](#toc0_)
Вы можете использовать UPDATE, если хотите заменить существующие значения в вашей базе данных на другие. Что делать, если, например, у кого-то в базе данных записаны рыжие волосы, хотя на самом деле у него каштановые волосы? Мы можем обновить эту запись с помощью инструкций UPDATE и WHERE.

```python
%%sql

UPDATE people_massachusetts
SET hair_color = 'brown'
WHERE first_name = 'Jane'
  AND last_name = 'Doe';
```

Или, скажем, в вашей таблице возникла проблема, из-за которой некоторые значения “address_state” отображаются как “Massachusetts”, а другие - как “MA”. Чтобы изменить все значения “MA” на “Massachusetts”, мы можем использовать простой запрос и обновить несколько записей одновременно.

```python
%%sql

UPDATE people_massachusetts
SET address_state = 'Massachusetts'
WHERE address_state = 'MA';
```

Будьте осторожны при использовании UPDATE. Если вы не укажете, какие записи следует изменить с помощью инструкции WHERE, вы измените все значения в таблице.

## <a id='toc1_20_'></a>[Используя команду DELETE](#toc0_)
УДАЛИТЬ, вы удалите записи из вашей таблицы. Как и в случае с UPDATE, обязательно укажите оператор WHERE, чтобы случайно не удалить всю таблицу целиком.

Или, если нам посчастливилось найти в нашей таблице people_massachusetts несколько записей, которые на самом деле проживали в штате Мэн, мы можем быстро удалить эти записи, указав на поле address_state.

```python
%%sql

DELETE
FROM people_massachusetts
WHERE address_state = 'maine';
```

## <a id='toc1_21_'></a>[Бонус: Дополнительные советы по SQL](#toc0_)
Теперь, когда вы научились создавать простые SQL-запросы, давайте обсудим некоторые другие приемы, которые вы можете использовать, чтобы повысить качество своих запросов, начиная со звездочки.

\* (asterisk)
Когда вы добавляете символ звездочки в свой SQL-запрос, он сообщает ему, что вы хотите включить все столбцы данных в свои результаты.

В примере с Массачусетсом, который мы использовали, у нас было только два имени столбцов: first_name и last_name. Но давайте предположим, что у нас есть 15 столбцов данных, которые мы хотим видеть в наших результатах — было бы непросто ввести все 15 имен столбцов в инструкции SELECT. Вместо этого, если вы замените названия этих столбцов звездочкой, запрос будет знать, как включить все столбцы в результаты.

Вот как будет выглядеть SQL-запрос.

```python
%%sql

SELECT *
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31'
ORDER BY last_name
LIMIT 100;
```

% (символ процента)
Символ процента является подстановочным знаком, то есть он может представлять один или несколько символов в значении базы данных. Подстановочные знаки полезны для поиска записей, в которых используются общие символы. Обычно они используются с оператором LIKE для поиска закономерности в данных.

Например, если мы хотим получить имена всех людей в нашей таблице, чей почтовый индекс начинается с “02”, мы можем написать следующий запрос.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE address_zip LIKE '02%';
```

Здесь “%” обозначает любую группу цифр, которые следуют за “02”, поэтому этот запрос выдает любую запись со значением для address_zip, которое начинается с “02”.

## <a id='toc1_22_'></a>[ЗА ПОСЛЕДНИЕ 30 ДНЕЙ](#toc0_)
Как только я начал регулярно использовать SQL, я обнаружил, что один из моих запросов к пользователям заключался в поиске того, какие люди совершали действия или удовлетворяли определенному набору критериев в течение последних 30 дней.

Давайте представим, что сегодня 1 декабря 2021 года. Вы могли бы создать эти параметры, указав дату рождения в интервале с 1 ноября 2021 года по 30 ноября 2021 года. Этот SQL-запрос будет выглядеть следующим образом:

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2021-11-01' AND '2021-11-30'
ORDER BY last_name
LIMIT 100;
```

Но для этого потребовалось бы учитывать, какие даты охватывают последние 30 дней, и вам пришлось бы постоянно обновлять этот запрос.

Вместо этого, чтобы даты автоматически охватывали последние 30 дней, независимо от того, какой это день, вы можете ввести это в поле И: birth_date >= (DATE_SUB(CURDATE(),ИНТЕРВАЛ 30))

(Примечание: Возможно, вам захочется еще раз проверить этот синтаксис у своей команды разработчиков, поскольку он может отличаться в зависимости от программного обеспечения, которое вы используете для обработки SQL-запросов.)

Таким образом, ваш полный SQL-запрос будет выглядеть следующим образом.

```python
%%sql

SELECT first_name,
       last_name
FROM people_massachusetts
WHERE hair_color = 'red'
  AND birth_date >= DATE_SUB(CURDATE(), INTERVAL 30 DAY)
ORDER BY last_name
LIMIT 100;
```

## <a id='toc1_23_'></a>[COUNT](#toc0_)
В некоторых случаях вам может потребоваться подсчитать, сколько раз появляется критерий поля. Например, предположим, вы хотите подсчитать, сколько раз разные цвета волос появлялись у людей из Массачусетса, которых вы оцениваете.

В этом случае пригодится функция подсчета, так что вам не придется вручную подсчитывать количество людей с разным цветом волос или экспортировать эту информацию в Excel.

Вот как будет выглядеть этот SQL-запрос:

```python
%%sql

SELECT hair_color,
       COUNT(hair_color)
FROM people_massachusetts
WHERE birth_date BETWEEN '2003-01-01' AND '2003-12-31'
GROUP BY hair_color;
```

## <a id='toc1_27_'></a>[С помощью команды JOIN](#toc0_)
Может возникнуть ситуация, когда вам потребуется получить доступ к информации из двух разных таблиц в одном SQL-запросе. В SQL для этого можно использовать предложение JOIN.

(Для тех, кто знаком с формулами Excel, это похоже на использование формулы ВПР, когда вам нужно объединить информацию из двух разных таблиц в Excel.)

Допустим, у нас есть одна таблица, в которой содержатся данные по идентификаторам пользователей и датам рождения всех жителей Массачусетса. Кроме того, у нас есть совершенно отдельная таблица, содержащая идентификаторы пользователей всех жителей Массачусетса и их цвет волос.

Если мы хотим определить цвет волос жителей Массачусетса, родившихся в 2003 году, нам нужно получить доступ к информации из обеих таблиц и объединить их. Это работает, потому что в обеих таблицах есть одинаковый столбец: идентификаторы пользователей.

Наша инструкция SELECT также немного изменится, поскольку мы будем вызывать поля из двух разных таблиц. Вместо простого перечисления полей, которые мы хотим включить в наши результаты, нам нужно будет указать, из какой таблицы они берутся.

(Примечание: Здесь может пригодиться функция asterisk, чтобы ваш запрос включал в результаты обе таблицы.)

Чтобы указать поле из конкретной таблицы, все, что нам нужно сделать, это объединить название таблицы с именем поля. Например, в нашей инструкции SELECT будет указано “таблица.поле” — с точкой, разделяющей названия таблицы и полей.

В этом случае мы также предполагаем несколько параметров:

1. Таблица даты рождения в Массачусетсе содержит следующие поля: first_name, last_name, user_id, дата рождения
2. Таблица цвета волос в Массачусетсе содержит следующие поля: user_id, hair_color

Ваш SQL-запрос будет выглядеть следующим образом.

```python
%%sql

SELECT birthdate_massachusetts.first_name,
       birthdate_massachusetts.last_name
FROM birthdate_massachusetts
JOIN haircolor_massachusetts USING (user_id)
WHERE hair_color = 'red'
  AND birth_date BETWEEN '2003-01-01' AND '2003-12-31'
ORDER BY last_name;
```

Этот запрос объединил бы две таблицы, используя поле “user_id”, которое отображается как в таблице birthdate_massachusetts, так и в таблице haircolor_massachusetts. Затем вы можете просмотреть таблицу людей с рыжими волосами, родившихся в 2003 году.

## <a id='toc1_28_'></a>[Используя оператор CASE](#toc0_)
Используйте оператор CASE, если вы хотите вернуть разные результаты в свой запрос в зависимости от того, какое условие выполнено. Условия оцениваются по порядку. Соответствующий результат возвращается при выполнении условия, а все последующие условия пропускаются.

Вы можете включить условие ELSE в конце, если не выполнено ни одно из условий. Без ELSE запрос вернет значение NULL, если не выполнено ни одно из условий.

Вот пример использования CASE для возврата строки на основе запроса.

```python
%%sql

SELECT first_name,
       last_name,
       CASE
         WHEN hair_color = 'brown' THEN 'This person has brown hair.'
         WHEN hair_color = 'blonde' THEN 'This person has blonde hair.'
         WHEN hair_color = 'red' THEN 'This person has red hair.'
         ELSE 'Hair color NOT known.'
       END AS hair_description
FROM people_massachusetts;
```
